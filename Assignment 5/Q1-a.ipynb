# Association Rule Mining for Books

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./book.csv')
df.info()
df.head()
```

```python
# build the rule
from mlxtend.frequent_patterns import apriori

rule = apriori(df, min_support=0.2, use_colnames=True)
rule
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
rule = association_rules(rule, metric='lift')
rule
```
