You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC). Build Linear Regression Model and plot regression line.

Dataset : IceCreamData.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./IceCreamData.csv')
df.head(10)
```

```python
df.cov()
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

### splitting the data into dependent and independent columns

```python
# y = Revenue
# y will always be a series (one dimensional array)
y = df['Revenue']

# x = Temperature
# x will always be a DataFrame (multidimensional)
x = df.drop('Revenue', axis=1)
```

### creating the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

### Result Visualization

```python
mean_value = np.ones(y.size) * y.mean()
mean_value
```

```python
plt.scatter(x['Temperature'], y, label='Observed Values')

plt.plot(x['Temperature'], mean_value)
plt.scatter(x['Temperature'], mean_value, label='mean value')

plt.plot(x['Temperature'], model.predict(x))
plt.scatter(x['Temperature'], model.predict(x), label='expected values')

plt.xlabel('Temperature')
plt.ylabel('Revenue')
plt.title('Linear Regression Model')
plt.legend()
```
