```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./movies.csv')
df.info()
df.head()
```

```python
df1=df.iloc[:,5:]
df1
```

```python
from mlxtend.preprocessing import TransactionEncoder

# create an encoder
encoder = TransactionEncoder()
# fit the transactions
encoder_transformed_data = encoder.fit(df1).transform(df1)
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
rule=apriori(df1, use_colnames=True, min_support=0.2)
rule
```

```python
rules = association_rules(rule,metric='lift')
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
rules[rules.lift>2]
```
