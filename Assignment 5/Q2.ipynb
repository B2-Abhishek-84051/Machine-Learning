### Create a Simple Linear Regression model for a given dataset. Also plot regression line.

### Datasets:
1. NewspaperData.csv
Independent = ‘Daily’, Dependent=’Sunday’

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./NewspaperData.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
covarince = np.cov(df['daily'], df['sunday'])
covarince
```

```python
df.drop('Newspaper', axis=1, inplace=True)
df.cov()
```

```python
## find the correlation between the columns to check the degree of relationship
```

```python
df.corr()
```

```python
import seaborn as sns
sns.heatmap(df.corr())
```

```python
sns.pairplot(df)
```

### Split the data into dependent and independent column

```python
y = df['sunday']
x = df.drop('sunday', axis=1)
```

#### create the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

#### result visualization

```python
plt.scatter(x['daily'], y, label='Observed Values', color="green") 


mean_values = np.ones(y.size)*y.mean()
plt.scatter(x, mean_values)
plt.plot(x, mean_values, label='mean values')

plt.scatter(x['daily'], model.predict(x))
plt.plot(x['daily'], model.predict(x), label='known values')

plt.title('Regression Line graph')
plt.xlabel('daily')
plt.ylabel('sunday')

plt.legend()
```

# Dataset
2. delivery_time.csv
Independent = ‘Sorting Time’, Dependent=’Delivery Time’

```python
df2 = pd.read_csv('./delivery_time.csv')
df2.info()
df2.head(10)
```

```python
df2.corr()
```

```python
df2.cov()
```

```python
import seaborn as sns
sns.heatmap(df2.corr())
```

```python
sns.pairplot(df2)
```

### split the data into dependent and independent columns

```python
y = df2['Delivery Time']
x = df2.drop('Delivery Time', axis=1)
```

```python
from sklearn.linear_model import LinearRegression

# create a new object of LinearRegression
model = LinearRegression()

# fit the data
model.fit(x, y)
```

### result visualization

```python
mean_values = np.ones(y.size) * y.mean()
mean_values
```

```python
# show the observed values
plt.scatter(x['Sorting Time'], y, label='Observed Values', color="green") 

# draw mean line
mean_values = np.ones(y.size) * y.mean()
plt.plot(x['Sorting Time'], mean_values, color='red')
plt.scatter(x['Sorting Time'], mean_values, color='red', label="mean values")

# draw the regression line
plt.plot(x['Sorting Time'], model.predict(x), color='blue')
plt.scatter(x['Sorting Time'], model.predict(x), color='blue', label="expected values")


# title
plt.title('Regression Line graph')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')

plt.legend()
```
