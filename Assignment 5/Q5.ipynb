In the following dataset you have given SAT score and GPA of students. You have to predict GPA of students by using SAT score.
Build a Linear Regression Model and plot regression line.

Dataset : SAT-GPA.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./SAT-GPA.csv')
df.head(10)
```

```python
df.cov()
```

```python
df.corr()
```

```python
sns.heatmap(df)
```

```python
sns.pairplot(df)
```

### splitting the data dependent and independent variables

```python
# y = dependent var - GPA
# y will always be a series (one dimensional array)
y = df['GPA']

# x = independent var - SAT
# x will always be DataFrame (multidimensional array)
x = df.drop('GPA', axis=1)
```

### create the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

### Result visualization

```python
mean_value = np.ones(y.size) * y.mean()
mean_value
```

```python
plt.scatter(x['SAT'], y, label='observed Values')

plt.plot(x['SAT'], mean_value)
plt.scatter(x['SAT'], mean_value, label='observed Values')

plt.plot(x['SAT'], model.predict(x))
plt.scatter(x['SAT'], model.predict(x), label='expected values')

plt.xlabel('SAT')
plt.ylabel('GPA')
plt.title('Linear Regression Model')
plt.legend()
```
