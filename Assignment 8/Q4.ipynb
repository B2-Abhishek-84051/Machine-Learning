```python
df.drop(['airborne','predator'], axis=1, inplace=True)
```

```python
# split the data into x and y
x = df.drop('type', axis = 1)
y = df['type']
```

```python
# split the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# train the model
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB

model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    return model_name, accuracy, precision, recall, f1
```

```python
res_knn = model_evaluation(model_knn, 'KNN')
res_svc = model_evaluation(model_svc, 'SVC')
res_nb = model_evaluation(model_nb, 'Naive Bayes')
```

```python
result_df = pd.DataFrame([res_knn, res_svc, res_nb], columns=['Model','Accuracy','Precision','Recall','F1'])
result_df
```

```python
y_pred = model_knn.predict(x_test)
plt.scatter(x_test['milk'][y_pred == 1], x_test['backbone'][y_pred == 1], color="green", label="Pass")
plt.scatter(x_test['milk'][y_pred == 0], x_test['backbone'][y_pred == 0], color="red", label="Fail")
plt.xlabel('milk')
plt.ylabel('backbone')
plt.title('Visualization for KNN')
plt.legend()
```
