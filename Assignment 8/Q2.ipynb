```python
# importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./framingham.csv')
df
```

```python
df.corr()
```

```python
x = df[['age','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','glucose']]
y = df['TenYearCHD']
```

```python
x.info()
```

```python
x['cigsPerDay'] = x['cigsPerDay'].fillna(x['cigsPerDay'].mean())
x['BPMeds'] = x['BPMeds'].fillna(x['BPMeds'].mean())
x['totChol'] = x['totChol'].fillna(x['totChol'].mean())
x['glucose'] = x['glucose'].fillna(x['glucose'].mean())
x['BMI'] = x['BMI'].fillna(x['BMI'].mean())
```

```python
# splitting the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# train the model
from sklearn.linear_model import LogisticRegressionCV

model_lg = LogisticRegressionCV(max_iter=1000)
model_lg.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
from sklearn.naive_bayes import GaussianNB

model_nb = GaussianNB()
model_nb.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# confusion matrix

def confusion_matrix_func(model):
    y_pred = model.predict(x_test)
    y_true = y_test
    return confusion_matrix(y_true, y_pred)
```

```python
print(f"confusion matrix for Logistic Regression = {confusion_matrix_func(model_lg)}")
print(f"confusion matrix for KNN = {confusion_matrix_func(model_knn)}")
print(f"confusion matrix for Support Vector Machine = {confusion_matrix_func(model_svc)}")
print(f"confusion matrix for Naive Bayes = {confusion_matrix_func(model_nb)}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    return model_name, accuracy, precision, recall, f1
```

```python
res_lg = model_evaluation(model_lg, 'Logistic Regression')
res_knn = model_evaluation(model_knn, 'KNN')
res_svc = model_evaluation(model_svc, 'SVC')
res_nb = model_evaluation(model_nb, 'Naive Bayes')
```

```python
result_df = pd.DataFrame([res_lg, res_knn, res_svc, res_nb], columns=['Model','Accuracy','Precision','Recall','F1'])
result_df
```

```python
def graph(model, model_name):
    y_pred = model.predict(x_test)
    plt.scatter(x_test['age'][y_pred == 1], x_test['prevalentHyp'][y_pred == 1], color="green", label="Pass")
    plt.scatter(x_test['age'][y_pred == 0], x_test['prevalentHyp'][y_pred == 0], color="red", label="Fail")
    plt.xlabel('age')
    plt.ylabel('prevalentHyp')
    plt.title(f'Visualization for {model_name}')
    plt.legend()
```

```python
graph(model_nb, 'Naive Bayes')
```
