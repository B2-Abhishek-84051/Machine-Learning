## Data Pre-processing

```python
df.drop(['Unnamed: 0'], axis=1 , inplace=True)
```

```python
df.columns
```

## Create DBScan Model

```python
from sklearn.cluster import DBSCAN

# create an object of DBScan
model = DBSCAN(eps=0.6, min_samples=12)

# fit the data
model.fit(df)
```

```python
model.labels_
```

```python
# create a figure using plt
figure = plt.figure()

# add a subplot
ax = figure.add_subplot(111)

colors = ['red', 'green', 'blue', 'magenta', 'yellow', "brown"]
for index in range(len(colors)):
    ax.scatter(df['Milk'][model.labels_ == index], df['Grocery'][model.labels_ == index], color=colors[index])

plt.xlabel('Income')
plt.ylabel('Spending Score')
```
