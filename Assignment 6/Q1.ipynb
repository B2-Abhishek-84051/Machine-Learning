```python
# importing required packages
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./ToyotaCorolla.csv', encoding='ISO-8859-1')
df.head(10)
```

```python
df.drop(['Model','Mfg_Month','Mfg_Year','Fuel_Type','Met_Color'], axis=1, inplace=True)                
```

```python
df.drop(['Color','Automatic','Cylinders','Mfr_Guarantee','BOVAG_Guarantee','Guarantee_Period','ABS','Airbag_1','Airbag_2','Airco','Automatic_airco','Boardcomputer','CD_Player','Central_Lock'], axis=1, inplace=True)
```

```python
df.drop(['Powered_Windows','Power_Steering','Radio','Mistlamps','Sport_Model','Backseat_Divider','Metallic_Rim','Radio_cassette','Tow_Bar'], axis=1, inplace=True)     
```

```python
df.head(10)
```

```python
df.drop('Id', axis=1, inplace=True)
df.head(10)
```
