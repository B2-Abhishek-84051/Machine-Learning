```python
import numpy as np
import pandas as pd
```

```python
# load the data
df = pd.read_csv('./framingham.csv')
df.info()
```

```python
df.corr()
```

```python
df.drop(['male','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','diabetes','totChol','BMI','heartRate'], axis=1, inplace=True)
df
```

```python
df.corr()
```

```python
# splitting the data into x and y
x = df.drop(['TenYearCHD'], axis=1)
y = df['TenYearCHD']
```

```python
# split the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.7, random_state=123456)
```

```python
# creating the model
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
```

```python
# evaluate the matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    return model_name, accuracy, precision, recall, f1
```

```python
result_df = pd.DataFrame([model_evaluation(model_dt, "Decision Tree"), model_evaluation(model_rf, "Random Forest")], columns=['Model','Accuracy','Precision','Recall score','F1 score'])
result_df
```
