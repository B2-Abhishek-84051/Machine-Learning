```python
import numpy as np
import pandas as pd
```

```python
# load the data
df = pd.read_csv("./glass.csv")
df.info()
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
df.drop(['K','Ca'], axis=1, inplace=True)
```

```python
# splitting the data into x and y
x = df.drop('Type', axis=1)
y = df['Type']
```

```python
y.unique()
```

```python
df['Type'].unique()
```

```python
df['Type'].value_counts()
```

```python
# splitting the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# create and train the model
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier(max_depth=3)
model_dt.fit(x_train, y_train)

from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=500, max_depth=6)
model_rf.fit(x_train, y_train)
```

```python
# model evaluation
from sklearn.metrics import confusion_matrix

def confusion_matrix_funct(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test

    print(f"Confusion Matrix for {model_name}= \n{confusion_matrix(y_true, y_pred)}")

confusion_matrix_funct(model_dt, "Decision Tree")
confusion_matrix_funct(model_rf, "Random Tree")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return model_name, accuracy, precision, recall, f1
```

```python
result_df = pd.DataFrame([model_evaluation(model_dt, "Decision Tree"), model_evaluation(model_rf, "Random Forest")], columns=['Model','Accuracy','Precision','Recall score','F1 score'])
result_df
```
