```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Churn.csv')
df
```

```python
df.info()
```

```python
# converting categorical column into numerical column
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Geography'] = encoder.fit_transform(df['Geography'])
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df.drop(['RowNumber','Surname','CustomerId'], axis = 1, inplace=True)
```

```python
df
```

```python
df.corr()
```

```python
# splitting the data into x and y
x = df[['Geography','Gender','Age','Balance','NumOfProducts','IsActiveMember']]
y = df['Exited']
```

```python
# splitting the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# Prepare model by using Decision Tree, Random Forest

from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
```

```python
# Calculate Accuracy by confusion matrix
from sklearn.metrics import confusion_matrix

def confusion_matrix_function(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test

    print(f"confusion matrix for {model_name} = {confusion_matrix(y_true, y_pred)}")
```

```python
confusion_matrix_function(model_dt, 'Decision Tree')
confusion_matrix_function(model_rf, 'Random Forest')
```

```python
# Calculate accuracy score, precision score, recall score, f1 score.
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    return model_name, accuracy, precision, recall, f1
```

```python
result_df = pd.DataFrame([model_evaluation(model_dt, "Decision Tree"), model_evaluation(model_rf, "Random Forest")], columns=['Model','accuracy','precision','recall score','F1 score'])
result_df
```
