```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./claimants.csv')
df
```

```python
df.info()
```

```python
df.corr()
```

```python
# splitting the data into x and y
x = df.drop(['CLMSEX','CLMINSUR','SEATBELT','CLMAGE','ATTORNEY'], axis=1)
y = df['ATTORNEY']
```

```python
x
```

```python
# splitting the data into test and train set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# creating and training model
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
# creating and training model
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=50, max_depth=6)
model_rf.fit(x_train, y_train)
```

```python
# creating the confusion matrix
from sklearn.metrics import confusion_matrix

def confusion_matrix_function(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    
    print(f"confusion matrix for {model_name} = {confusion_matrix(y_true, y_pred)}")

confusion_matrix_function(model_dt, 'Decision Tree')
confusion_matrix_function(model_rf, 'Random Forest')
```

```python
# Calculate accuracy score, precision score, recall score, f1 score.
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def model_evaluation(model, model_name):
    y_pred = model.predict(x_test)
    y_true = y_test
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return model_name, accuracy, precision, recall, f1
```

```python
result_df = pd.DataFrame([model_evaluation(model_dt, "Decision Tree"), model_evaluation(model_rf, "Random Forest")], columns=['Model','Accuracy','Precision','Recall score','F1 score'])
result_df
```
