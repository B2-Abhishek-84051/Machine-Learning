```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('framingham.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['education'] = df['education'].fillna(df['education'].mode()[0])
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0])
df['BPMeds'] = df['BPMeds'].fillna(df['BPMeds'].mode()[0])
df['totChol'] = df['totChol'].fillna(df['totChol'].mean())
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mean())
df['glucose'] = df['glucose'].fillna(df['glucose'].mean())
```

```python
df.info()
```

```python
# splitting the data into x and y
x = df.drop(['TenYearCHD'], axis=1)
y = df['TenYearCHD']
```

```python
# splitting the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# creating the model
from sklearn.linear_model import LogisticRegressionCV

model = LogisticRegressionCV(cv=2, max_iter=10000)
model.fit(x_train, y_train)
```

```python
# model evaluation
y_pred = model.predict(x_test)
y_true = y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print(f"confusion matrix = {confusion_matrix(y_true, y_pred)}")
print(f"accuracy score = {accuracy_score(y_true, y_pred)*100:0.2f}%")
print(f"precision score = {precision_score(y_true, y_pred)*100:0.2f}%")
print(f"recall score = {recall_score(y_true, y_pred)}")
print(f"F1 score = {f1_score(y_true, y_pred)*100:0.2f}%")
print(f"roc auc score = {roc_auc_score(y_true, y_pred)*100:0.2f}%")
```
