In this dataset you have to predict that whether we have to give ATTORNEY to the clients or not (0=Yes, 1=No) based on below factors.
CLMAGE - Age of Client
CLMSEX - Sex of client (0=Male, 1= Female)
SEATBELT - Whether client has Seatbelt or not (0=Yes, 1=No)
CLMINSUR - Whether Client give Insurance or not (0=Yes, 1=No)
LOSS - Loss that client faced

Prepare model by using logistic regression & Also calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score, roc auc score.

Dataset : Claimants.csv

```python
# importing the required packages
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# load the data into DataFrame
df = pd.read_csv('./claimants.csv')
df.head(10)
```

```python
df.info()
```

```python
# df['CLMSEX'] = df.fillna()
df['CLMSEX'] = df['CLMSEX'].fillna(df['CLMSEX'].mode()[0])
df['CLMINSUR'] = df['CLMINSUR'].fillna(df['CLMINSUR'].mode()[0])
df['SEATBELT'] = df['SEATBELT'].fillna(df['SEATBELT'].mode()[0])
df['CLMAGE'] = df['CLMAGE'].fillna(df['CLMAGE'].mean())
```

```python
df.info()
```

```python
# splitting the data into x and y
x = df.drop(['ATTORNEY'], axis=1)
y = df['ATTORNEY']
```

```python
# splitting the data into train and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=123456)
```

```python
# build the model
from sklearn.linear_model import LogisticRegressionCV

model = LogisticRegressionCV(cv=2)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_true = y_test
```

```python
# model evaluation
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print(f"confusion metrix = {confusion_matrix(y_true, y_pred)}")
print(f"accuracy score = {accuracy_score(y_true, y_pred)*100:0.2f}%")
print(f"precision score = {precision_score(y_true, y_pred)*100:0.2f}%")
print(f"recall score = {recall_score(y_true, y_pred)}")
print(f"F1 score = {f1_score(y_true, y_pred)*100:0.2f}%")
print(f"roc auc score = {roc_auc_score(y_true, y_pred)*100:0.2f}%")
```
