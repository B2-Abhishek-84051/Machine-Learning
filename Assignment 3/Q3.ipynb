```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./universities_std.csv')
df
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0','Top10','Accept','SFRatio','Expenses'], axis=1, inplace=True)
df
```

```python
no_of_clusters = np.arange(1, 11)
wss = []

for k in no_of_clusters:
    model = KMeans(n_clusters=k, random_state=12345)
    model.fit(df)
    wss.append(model.inertia_)
```

```python
plt.plot(no_of_clusters, wss, label='wss')
plt.scatter(no_of_clusters, wss)
plt.xlabel('No of cluster')
plt.ylabel('Within sum of squares')
plt.title('Elbow Menthd Visualization')
plt.legend()
```

```python
model = KMeans(n_clusters=6, random_state = 123)
model.fit(df)
```

```python
plt.xlabel("SAT")
plt.ylabel("Grade Score")
plt.title("University")

plt.scatter(
    df['SAT'][model.labels_ == 0],
    df['GradRate'][model.labels_ == 0],
    color="red")
plt.scatter(
    df['SAT'][model.labels_ == 1],
    df['GradRate'][model.labels_ == 1],
    color="g")
```
